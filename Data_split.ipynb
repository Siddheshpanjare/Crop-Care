```python
#Augmentated data splitting
import os
import shutil
from sklearn.model_selection import train_test_split

# Original dataset path
dataset_path = r"F:\Project\CropDiseases(Aug)\CropDiseases\CropDiseases\Crop___Disease"

# Output folder
output_path = r"F:\Project\CropDiseases(Aug)\CropDiseases"

# Allowed image extensions
valid_extensions = ('.jpg', '.jpeg', '.png')

# Loop through main crop folders
for crop in os.listdir(dataset_path):

    crop_path = os.path.join(dataset_path, crop)

    if not os.path.isdir(crop_path):
        continue

    # Loop through disease folders
    for disease in os.listdir(crop_path):

        disease_path = os.path.join(crop_path, disease)

        if not os.path.isdir(disease_path):
            continue

        # Get only image files
        images = [
            img for img in os.listdir(disease_path)
            if img.lower().endswith(valid_extensions)
        ]

        # Split dataset
        train_imgs, temp_imgs = train_test_split(
            images,
            test_size=0.30,
            random_state=42
        )

        val_imgs, test_imgs = train_test_split(
            temp_imgs,
            test_size=0.50,
            random_state=42
        )

        splits = {
            "train": train_imgs,
            "val": val_imgs,
            "test": test_imgs
        }

        # Copy files
        for split_name, split_imgs in splits.items():

            split_dir = os.path.join(output_path, split_name, disease)
            os.makedirs(split_dir, exist_ok=True)

            for img in split_imgs:

                src = os.path.join(disease_path, img)
                dst = os.path.join(split_dir, img)

                shutil.copy2(src, dst)

print("Dataset successfully split!")
```

```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ============================================
# TRAIN DIRECTORY
# ============================================

TRAIN_DIR = r"F:\Project\CropDiseases(Aug)\CropDiseases\split_dataset\train"

# ============================================
# CLASSES TO AUGMENT
# ============================================

TARGET_CLASSES = [
    "Rice_Bacterial leaf blight",
    "Rice_Leaf_smut"
]

# ============================================
# HOW MANY NEW IMAGES PER ORIGINAL
# ============================================

AUGMENTATIONS_PER_IMAGE = 3

# ============================================
# AUGMENTATION PIPELINE
# ============================================

datagen = ImageDataGenerator(

    rotation_range=25,

    width_shift_range=0.15,
    height_shift_range=0.15,

    zoom_range=0.2,

    shear_range=0.15,

    horizontal_flip=True,

    brightness_range=[0.8, 1.2],

    fill_mode='nearest'
)

# ============================================
# GENERATE AUGMENTED IMAGES
# ============================================

for class_name in TARGET_CLASSES:

    class_path = os.path.join(TRAIN_DIR, class_name)

    print(f"\nProcessing Class: {class_name}")

    images = os.listdir(class_path)

    count = 0

    for img_name in images:

        img_path = os.path.join(class_path, img_name)

        # Read image
        image = cv2.imread(img_path)

        if image is None:
            continue

        # Convert BGR to RGB
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Reshape for generator
        image = np.expand_dims(image, axis=0)

        # Generate augmented images
        aug_iter = datagen.flow(
            image,
            batch_size=1
        )

        for i in range(AUGMENTATIONS_PER_IMAGE):

            aug_image = next(aug_iter)[0].astype(np.uint8)

            save_name = f"aug_{count}_{i}.jpg"

            save_path = os.path.join(
                class_path,
                save_name
            )

            # Convert RGB back to BGR
            aug_image = cv2.cvtColor(
                aug_image,
                cv2.COLOR_RGB2BGR
            )

            cv2.imwrite(save_path, aug_image)

        count += 1

    print(f"Added augmented images to: {class_name}")

print("\nOffline augmentation completed.")
```
